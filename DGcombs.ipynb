```python
from tqdm.notebook import tqdm
from math import comb
import pandas as pd

steps = 10
combos = [
    (
        1.0 * a / steps,
        1.0 * b / steps,
        1.0 * c / steps,
        1.0 * d / steps,
        1.0 * e / steps,
        1.0 * f / steps,
        1.0 * g / steps,
        1.0 * h / steps,
        1.0 * i / steps,
        1.0 * j / steps,
    )
    for a in tqdm(range(steps + 1))
    for b in range(steps + 1 - a)
    for c in range(steps + 1 - a - b)
    for d in range(steps + 1 - a - b - c)
    for e in range(steps + 1 - a - b - c - d)
    for f in range(steps + 1 - a - b - c - d - e)
    for g in range(steps + 1 - a - b - c - d - e - f)
    for h in range(steps + 1 - a - b - c - d - e - f - g)
    for i in range(steps + 1 - a - b - c - d - e - f - g - h)
    for j in range(steps + 1 - a - b - c - d - e - f - g - h - i)
    if a + b + c + d + e + f + g + h + i + j == steps
    and a >= 0
    and b >= 0
    and c >= 0
    and d >= 0
    and e >= 0
    and f >= 0
    and g >= 0
    and h >= 0
    and i >= 0
    and j >= 0
]
csys = tuple("ABCDEFGHIJ")
comps = [{k: v for k, v in zip(csys, combo) if v > 0} for combo in combos]

comps_per_space = len(comps)
num_spaces = comb(27, 10)
totals = comps_per_space * num_spaces

comps
```

```python
# generate the posssible combinations of 4 components from 27
# by filling a 4x27 matrix with 1s and 0s

import numpy as np
from itertools import combinations

n = 27
k = 4
combinations = list(combinations(range(n), k))

count=len(combinations)

# iterate over the combinations and check if three numbers are the same regardless of order
# if so then subtract 1 from the count

# do this by taking two rows and converting them to a set
# then taking the difference of the two sets
# if the len(difference) is 1 then the two rows contain the same numbers except for one number
# in this case subtract 1 from the count

#count is a deep copy of count 1
sub=0
for i in range(count):
    for j in range(i+1,count):
        if len(set(combinations[i])-set(combinations[j]))==1:
            sub-=1

sub
```
